```python
import json
import textstat
from textatistic import Textatistic # http://www.erinhengel.com/software/textatistic/
import textatistic
from unidecode import unidecode
import csv

minMedianIncome = 9931
maxMedianIncome = 270995
incomeDenominator = maxMedianIncome - minMedianIncome

minTravelTime = 23
maxTravelTime = 63
travelTimeDenominator = maxTravelTime - minTravelTime

minRent = 220
maxRent = 2734
rentDenominator = maxRent - minRent

minHousingUnits = 39682
maxHousingUnits = 148515
housingUnitsDenominator = maxHousingUnits - minHousingUnits

minDiversityIndex = 0.17
maxDiversityIndex = 0.83
diversityIndexDenominator = maxDiversityIndex - minDiversityIndex

minPopulation = 98174
maxPopulation = 264586
populationDenominator = maxPopulation - minPopulation

minPovertyRate = 2.2
maxPovertyRate = 52.2
povertyDenominator = maxPovertyRate = minPovertyRate


# we want an array of objects where each object represents a story
everything = {}

# Get SubBorough Names
with open("./raw_data/median_household_income.csv") as f:
    readCSV = csv.reader(f, delimiter=',')
    for index,row in enumerate(readCSV):      # each row will represent another city
        if(index != 0):
            everything[row[2]] = {}    #row[2] is the name of the sub-borough

years = [2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016]

with open("./raw_data/median_household_income.csv") as f:
    readCSV = csv.reader(f, delimiter=',')

    for index,row in enumerate(readCSV):      # each row will represent another city
        if(index != 0):
            everything[row[2]]["median_household_income"] = {}
            everything[row[2]]["median_household_income"]["actual"] = []
            everything[row[2]]["median_household_income"]["normalized"] = []
            
            start = 3
            for year in years:
                actual = {}
                normalized = {}
                actual["year"] = year
                normalized["year"] = year
                
                actual["median_household_income"] = int(float(row[start]))
                normalized["median_household_income"] = round((int(float(row[start])) - minMedianIncome)/incomeDenominator,2)
                
                everything[row[2]]["median_household_income"]["actual"].append(actual)
                everything[row[2]]["median_household_income"]["normalized"].append(normalized)

                start = start + 1
                
                
with open("./raw_data/mean_travel_time_to_work.csv") as f:
    readCSV = csv.reader(f, delimiter=',')

    for index,row in enumerate(readCSV):      # each row will represent another city
        if(index != 0):
            everything[row[2]]["mean_travel_time"] = {}
            everything[row[2]]["mean_travel_time"]["actual"] = []
            everything[row[2]]["mean_travel_time"]["normalized"] = []
            
            start = 3
            for year in years:
                actual = {}
                normalized = {}
                actual["year"] = year
                normalized["year"] = year
                if(row[start]):
                    actual["mean_travel_time"] = int(float(row[start]))
                    normalized["mean_travel_time"] = round(1 - round((int(float(row[start])) - minTravelTime)/travelTimeDenominator,2),2)
                else:
                    actual["mean_travel_time"] = "None"
                    normalized["mean_travel_time"] = "None"
                everything[row[2]]["mean_travel_time"]["actual"].append(actual)
                everything[row[2]]["mean_travel_time"]["normalized"].append(normalized)

                start = start + 1
                
                
with open("./raw_data/median_rent.csv") as f:
    readCSV = csv.reader(f, delimiter=',')

    for index,row in enumerate(readCSV):      # each row will represent another city
        if(index != 0):
            everything[row[2]]["median_rent"] = {}
            everything[row[2]]["median_rent"]["actual"] = []
            everything[row[2]]["median_rent"]["normalized"] = []
            
            start = 3
            for year in years:
                actual = {}
                normalized = {}
                actual["year"] = year
                normalized["year"] = year
                if(row[start]):
                    actual["median_rent"] = int(float(row[start]))
                    normalized["median_rent"] = round((int(float(row[start])) - minRent)/rentDenominator,2)
                else:
                    actual["median_rent"] = "None"
                    normalized["median_rent"] = "None"
                everything[row[2]]["median_rent"]["actual"].append(actual)
                everything[row[2]]["median_rent"]["normalized"].append(normalized)

                start = start + 1
            
            
with open("./raw_data/housing_units.csv") as f:
    readCSV = csv.reader(f, delimiter=',')

    for index,row in enumerate(readCSV):      # each row will represent another city
        if(index != 0):
            everything[row[2]]["housing_units"] = {}
            everything[row[2]]["housing_units"]["actual"] = []
            everything[row[2]]["housing_units"]["normalized"] = []
            
            start1 = 3
            for year in years: 
                actual = {}
                normalized = {}
                actual["year"] = year
                normalized["year"] = year
                if(row[start1]):
                    actual["housing_units"] = int(float(row[start1]))
                    normalized["housing_units"] = round((int(float(row[start1])) - minHousingUnits)/housingUnitsDenominator,2)
                else:
                    actual["housing_units"] = "None"
                    normalized["housing_units"] = "None"
                everything[row[2]]["housing_units"]["actual"].append(actual)
                everything[row[2]]["housing_units"]["normalized"].append(normalized)

                start1 = start1 + 1
                
                
with open("./raw_data/racial_diversity_index.csv") as f:
    readCSV = csv.reader(f, delimiter=',')

    for index,row in enumerate(readCSV):      # each row will represent another city
        if(index != 0):
            everything[row[2]]["racial_diversity_index"] = {}
            everything[row[2]]["racial_diversity_index"]["actual"] = []
            everything[row[2]]["racial_diversity_index"]["normalized"] = []
            
            start1 = 3
            for year in years: 
                actual = {}
                normalized = {}
                actual["year"] = year
                normalized["year"] = year
                if(row[start1]):
                    actual["racial_diversity_index"] = round(float(row[start1]),2)
                    
                    # Here, we take the compliment. That way a higher normalized racial diversity index implies lower diversity!
                    normalized["racial_diversity_index"] = round(1 - round(((float(row[start1])) - minDiversityIndex)/diversityIndexDenominator,2),2)
                else:
                    actual["racial_diversity_index"] = "None"
                    normalized["racial_diversity_index"] = "None"
                everything[row[2]]["racial_diversity_index"]["actual"].append(actual)
                everything[row[2]]["racial_diversity_index"]["normalized"].append(normalized)

                start1 = start1 + 1
                
                
                
with open("./raw_data/population.csv") as f:
    readCSV = csv.reader(f, delimiter=',')

    for index,row in enumerate(readCSV):      # each row will represent another city
        if(index != 0):
            everything[row[2]]["population"] = {}
            everything[row[2]]["population"]["actual"] = []
            everything[row[2]]["population"]["normalized"] = []
            
            start1 = 3
            for year in years: 
                actual = {}
                normalized = {}
                actual["year"] = year
                normalized["year"] = year
                if(row[start1]):
                    actual["population"] = int(float(row[start1]))
                    normalized["population"] = round((int(float(row[start1])) - minPopulation)/populationDenominator,2)
                else:
                    actual["population"] = "None"
                    normalized["population"] = "None"
                everything[row[2]]["population"]["actual"].append(actual)
                everything[row[2]]["population"]["normalized"].append(normalized)

                start1 = start1 + 1

                
                
#with open("./raw_data/population.csv") as f:
#    readCSV = csv.reader(f, delimiter=',')
#    for index,row in enumerate(readCSV):      # each row will represent another city
#        if(index != 0):
#            everything[row[2]]["population"] = []
#            start = 3
#            for year in years:
#                kebab = {}
#                kebab["year"] = year
#                kebab["population"] = int(float(row[start]))
#                everything[row[2]]["population"].append(kebab)

#                start = start + 1


# Definition: The probability that two randomly chosen people in a given geographic area will be of a different race.
#with open("./raw_data/racial_diversity_index.csv") as f:  #
#    readCSV = csv.reader(f, delimiter=',')

#    for index,row in enumerate(readCSV):      # each row will represent another city
#        if(index != 0):
#            start = 3
#            everything[row[2]]["racial_diversity_index"] = []
#            for year in years:
#                kebab1 = {}
#                kebab1["year"] = year
#                if(row[start]):
#                    kebab1["racial_diversity_index"] = round(float(row[start])*100,1)
#                    
#                else:
#                    kebab1["racial_diversity_index"] = "none"
#                everything[row[2]]["racial_diversity_index"].append(kebab1)
#                start = start + 1

            
            
            
# Definition: The number of people below the poverty threshold divided by the number of people for whom poverty status was determined.
#with open("./raw_data/poverty_rate.csv") as f:  #
#    readCSV = csv.reader(f, delimiter=',')

#    for index,row in enumerate(readCSV):      # each row will represent another city
#        if(index != 0):
#            everything[row[2]]["poverty_rate"] = []
#            start = 3
#            for year in years:
#                kebab = {}
#                kebab["year"] = year
#                if(row[start]):
#                    kebab["poverty_rate"] = round(float(row[start])*100,1)
#                else:
#                    kebab["poverty_rate"] = "none"
#                start = start + 1
#                everything[row[2]]["poverty_rate"].append(kebab)


#with open("./raw_data/unemployment_rate.csv") as f:  #
#    readCSV = csv.reader(f, delimiter=',')

#    for index,row in enumerate(readCSV):      # each row will represent another city
#        if(index != 0):
#            everything[row[2]]["unemployment_rate"] = []
#            start = 3
#            for year in years:
#                kebab = {}
#                kebab["year"] = year
#                if(row[start]):
#                    kebab["unemployment_rate"] = round(float(row[start])*100,1)
#                else:
#                    kebab["unemployment_rate"] = "none"
#                start = start + 1
#                everything[row[2]]["unemployment_rate"].append(kebab)

                
#with open("./raw_data/percent_white.csv") as f:  #
#    readCSV = csv.reader(f, delimiter=',')
#
#    for index,row in enumerate(readCSV):      # each row will represent another city
#        if(index != 0):
#            start = 3
#            everything[row[2]]["percent_white"] = []
#            for year in years:
#                kebab = {}
#                kebab["year"] = year
#                if(row[start]):
#                    kebab["percent_white"] = round(float(row[start])*100,1)
#                else:
#                    kebab["percent_white"] = "none"
#                start = start + 1
#                everything[row[2]]["percent_white"].append(kebab)


#with open("./raw_data/percent_black.csv") as f:  #
#    readCSV = csv.reader(f, delimiter=',')
#
#    for index,row in enumerate(readCSV):      # each row will represent another city
#        if(index != 0):
#            everything[row[2]]["percent_black"] = []
#            start = 3
#            for year in years:
#                kebab = {}
#                kebab["year"] = year
#                if(row[start]):
#                    kebab["percent_black"] = round(float(row[start])*100,1)
#                else:
#                    kebab["percent_black"] = "none"
#                start = start + 1
#                everything[row[2]]["percent_black"].append(kebab)

                
#with open("./raw_data/percent_hispanic.csv") as f:  #
#    readCSV = csv.reader(f, delimiter=',')
#
#    for index,row in enumerate(readCSV):      # each row will represent another city
#        if(index != 0):
#            everything[row[2]]["percent_hispanic"] = []
#            start = 3
#            for year in years:
#                kebab = {}
#                kebab["year"] = year
#                if(row[start]):
#                    kebab["percent_hispanic"] = round(float(row[start])*100,1)
#                else:
#                    kebab["percent_hispanic"] = "none"
#                start = start + 1
#                everything[row[2]]["percent_hispanic"].append(kebab)


#with open("./raw_data/percent_asian.csv") as f:  #
#    readCSV = csv.reader(f, delimiter=',')
#
#    for index,row in enumerate(readCSV):      # each row will represent another city
#        if(index != 0):
#            everything[row[2]]["percent_asian"] = []
#            start = 3
#            for year in years:
#                kebab = {}
#                kebab["year"] = year
#                if(row[start]):
#                    kebab["percent_asian"] = round(float(row[start])*100,1)
#                else:
#                    kebab["percent_asian"] = "none"
#                start = start + 1
#                everything[row[2]]["percent_asian"].append(kebab)

            
            
            
  
with open('data.json', 'w') as outfile:
    json.dump(everything, outfile)
    
```
